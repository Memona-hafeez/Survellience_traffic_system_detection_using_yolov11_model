```python
from ultralytics import YOLO

# Load YOLOv8 model (not OBB)
model = YOLO('yolov8s.pt')

# Train model
model.train(
    data='/content/UA-DETRAC-DATASET-10K-2/data.yaml',
    epochs=20,
    imgsz=640,
    batch=16,
    device=0,
    optimizer='SGD',
    lr0=0.01,
    project='UA-DETRAC_yolov8_training',
    name='mma_detector_v8',
    exist_ok=True,
    pretrained=True,
    mosaic=True,
    mixup=True,
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=2,
)

# Load trained model for inference
trained_model_path = 'UA-DETRAC_yolov8_training/mma_detector_v8/weights/best.pt'
trained_model = YOLO(trained_model_path)
```

```python
from google.colab import files
model_path = "UA-DETRAC_yolov8_training/mma_detector_v8/weights/best.pt"
files.download(model_path)
```

```python
# Run validation
metrics = model.val()

# Show key metrics
print(f"Precision: {metrics.box.p[0]:.4f}")  # Access precision for class 0
print(f"Recall: {metrics.box.r[0]:.4f}")      # Access recall for class 0
print(f"mAP@0.5: {metrics.box.map50:.4f}")  # Access mAP@0.5
print(f"mAP@0.5:0.95: {metrics.box.map:.4f}") # Access mAP@0.5:0.95
```
